Question 3: What is clustering in unsupervised learning? Mention three popular
clustering algorithms.

 =Clustering is an unsupervised learning technique that groups similar data points together based on their features, without using any predefined labels. The goal is to ensure that data points within the same cluster are more similar to each other than to those in other clusters.

Three popular clustering algorithms:

K-Means Clustering – Divides data into k clusters based on distance to cluster centroids.

Hierarchical Clustering – Builds a tree-like structure (dendrogram) to group data points at different levels.

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) – Groups data points based on density and identifies outliers as noise.


Question 4: Explain the concept of anomaly detection and its significance.

 =Anomaly detection is the process of identifying data points or patterns that deviate significantly from normal behavior. These unusual observations are called anomalies or outliers.

Significance:

Helps detect fraudulent activities (e.g., credit card fraud).

Identifies network intrusions or cyberattacks.

Detects faults in machines or defects in manufacturing.

Improves data quality by removing incorrect or rare values.


Question 5: List and briefly describe three types of anomaly detection techniques.

 =Statistical Methods:
Identify anomalies by assuming normal data follows a specific distribution (e.g., Gaussian) and flagging points that lie far from the mean.

Machine Learning–Based Methods:
Use models like Isolation Forest, One-Class SVM, or Autoencoders to learn normal patterns and detect deviations.

Distance or Density-Based Methods:
Detect anomalies by measuring how far a point is from others (e.g., k-NN) or by checking if it lies in a low-density region (e.g., DBSCAN).


Question 6: What is time series analysis? Mention two key components of time series
data.

 =Time series analysis is the study of data points collected or recorded over time to identify trends, patterns, and seasonal variations for forecasting or decision-making.

Two key components of time series data:

Trend: The long-term upward or downward movement in the data over time.

Seasonality: Regular and repeating patterns or fluctuations that occur at specific intervals


Question 7: Describe the difference between seasonality and cyclic behavior in time
series.

 =Seasonality refers to regular and predictable patterns that repeat over a fixed period, such as daily, monthly, or yearly (e.g., increased ice cream sales every summer).
Cyclic behavior, on the other hand, refers to fluctuations that occur over irregular and longer periods, often influenced by economic or business cycles (e.g., market growth and recession phases).
In short:


Seasonality → fixed, repeating intervals.


Cyclic behavior → irregular, long-term fluctuations.





```python
#Question 8: Write Python code to perform K-means clustering on a sample dataset.

# Import necessary libraries
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Create a sample dataset
X, y = make_blobs(n_samples=200, centers=3, random_state=42, cluster_std=1.2)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Get cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title("K-Means Clustering Example")
plt.legend()
plt.show()
```

Question 9: What is inheritance in OOP? Provide a simple example in Python.

 =Inheritance in Object-Oriented Programming (OOP) is a mechanism that allows one class (child or subclass) to acquire the properties and methods of another class (parent or superclass). It promotes code reusability and extensibility.

```python
# Parent class
class Animal:
    def speak(self):
        print("This animal makes a sound.")

# Child class
class Dog(Animal):
    def speak(self):
        print("The dog barks.")

# Create objects
a = Animal()
d = Dog()

a.speak()  # Output: This animal makes a sound.
d.speak()  # Output: The dog barks.
```
